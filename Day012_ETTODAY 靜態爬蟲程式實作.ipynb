### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import re
from datetime import date

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
r.encoding = 'utf-8'
#print(r.text)

soup = BeautifulSoup(r.text,'html.parser')
#print(soup)
article_date = []
category = []
title = []
for h in soup.find(class_="part_list_2").find_all('h3'):
    article_date.append(h.find(class_="date").text)
    category.append(h.find('em').text)
    title.append(h.find('a').text)
    #print(article_date)
    #print(category)
    #print(title)

#print(article_date)
#print(category)
#print(title)

#print(len(article_date))

todaydate = str(date.today())
todaydate = todaydate.replace('-','/')
#print(todaydate)

for a in zip(article_date,category,title):
    
    pattern = re.compile(todaydate+'\s{1}\d{2}:\d{2}')
    regex_match = re.search(pattern,a[0])
    #print(regex_match)
    if regex_match != None:
        print(a[0],a[2])
    
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python

new_category = []
new_title = []

for i in zip(article_date,category,title):
    
    if i[1] in new_category:
        #print(new_category.index(i[1]))
        new_title[new_category.index(i[1])].append(i[2])
        
    else:
        new_category.append(i[1])
        category_title = []
        category_title.append(i[2])
        new_title.append(category_title)
        #print(new_category)
        #print(new_title)

#print(new_category)
#print('-----------------')
#print(new_title)
#print('-----------------')
i = dict(zip(new_category,new_title))
print(i)

    
```

### ③ 哪一個類別的文章最多

```python
count = []
for i in range(len(new_title)):
    count.append(len(new_title[i]))
print(count)
#print(new_category)
print('文章最多的類別：',new_category[count.index(max(count))])
#print(len(new_category))
```
