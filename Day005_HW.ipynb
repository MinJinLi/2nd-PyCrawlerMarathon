## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(r.text)
print(type(data))
print('一次回傳{}筆資料'.format(len(data)))
print('每一筆資料包含：')
for i in data[0].keys():
    print(i)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for i in range(len(data)):
    print('標題:',data[i]['title'])
    print('貼文時間:',data[i]['createdAt'])
    print('留言人數:',data[i]['commentCount'])
    print('按讚人數:',data[i]['likeCount'],'\n')
```

```python
D = []
for i in range(len(data)):
    element = []
    element.append(data[i]['title'])
    element.append(data[i]['createdAt'])
    element.append(data[i]['commentCount'])
    element.append(data[i]['likeCount'])

    D.append(element)

print(D)
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

#熱門
comment = 0
like = 0
for i in range(len(data)):
    comment += data[i]['commentCount'] 
    like += data[i]['likeCount']

print('平均留言人數:', comment/len(data))
print('平均按讚人數:', like/len(data))
```

```python
#非熱門
no_r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
no_j = json.loads(no_r.text)
#print(no_j)

no_comment = 0
no_like = 0
for i in range(len(no_j)):
    no_comment += no_j[i]['commentCount'] 
    no_like += no_j[i]['likeCount']

print('平均留言人數:', no_comment/len(no_j))
print('平均按讚人數:', no_like/len(no_j))
```
