```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(r.text)
print(type(data))
print('ä¸€æ¬¡å›å‚³{}ç­†è³‡æ–™'.format(len(data)))
print('æ¯ä¸€ç­†è³‡æ–™åŒ…å«ï¼š')
for i in data[0].keys():
    print(i)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for i in range(len(data)):
    print('æ¨™é¡Œ:',data[i]['title'])
    print('è²¼æ–‡æ™‚é–“:',data[i]['createdAt'])
    print('ç•™è¨€äººæ•¸:',data[i]['commentCount'])
    print('æŒ‰è®šäººæ•¸:',data[i]['likeCount'],'\n')
```

```python
D = []
for i in range(len(data)):
    element = []
    element.append(data[i]['title'])
    element.append(data[i]['createdAt'])
    element.append(data[i]['commentCount'])
    element.append(data[i]['likeCount'])

    D.append(element)

print(D)
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

#ç†±é–€
comment = 0
like = 0
for i in range(len(data)):
    comment += data[i]['commentCount'] 
    like += data[i]['likeCount']

print('å¹³å‡ç•™è¨€äººæ•¸:', comment/len(data))
print('å¹³å‡æŒ‰è®šäººæ•¸:', like/len(data))
```
